```python
df['sex_encoded']=df['sex'].map({'male':0,'female':1})
```

```python
print(df.dtypes)
```

```python
non_numeric = df.select_dtypes(exclude='number').columns
print("Non-numeric columns:",non_numeric)
```

```python
df.head()
```

```python
# Encoding multiple columns
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
df['class'] = df['class'].map({'Third': 0, 'Second': 1, 'First': 2})
df['who'] = df['who'].map({'man': 0, 'woman': 1, 'child': 2})
df['alive'] = df['alive'].map({'no': 0, 'yes': 1})
df['alone'] = df['alone'].astype(int)           # True/False to 1/0
df['adult_male'] = df['adult_male'].astype(int) # True/False to 1/0
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
```

```python
#Survival Count
sns.countplot(x='survived',data=df)
plt.title("Survival Count (0 = Died , 1 = Survived)")
plt.xlabel("Survived")
plt.ylabel("Passenger Count")
plt.show()
```

```python
#Survival by gender
sns.countplot(x='sex',hue='survived',data=df)
plt.title('Survival by Gender')
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title="Survived",labels=['No','Yes'])
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x='survived',y='age',hue='sex',data=df)
plt.title("Age vs Survival (by Gender)")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.legend(title='Sex')
plt.show()
```

```python
#Violin plot - Age Distribution by class and gender
sns.violinplot(x='pclass',y='age',hue='sex',data=df,split=True)
plt.title("Age Distribution by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```
