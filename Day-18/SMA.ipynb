```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

.rolling(window=5) : Takes 5-day window

```python
#Simulated data
data = {'Day':pd.date_range(start='2025-01-01',periods=10),
'Sales':[200,220,210,230,240,250,260,270,265,275]}
df  = pd.DataFrame(data)
df['SMA_5'] = df['Sales'].rolling(window=5).mean()

print(df)
```

```python
plt.plot(df['Day'],df['Sales'],label='Original')
plt.plot(df['Day'],df['SMA_5'],label='5-Day SMA',color='orange')
plt.legend()
plt.title("Simple Moving Average")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
