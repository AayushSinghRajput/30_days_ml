```python
#Import Required Libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#Load the Iris Dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data,columns=iris.feature_names) #Create a DataFrame from the feature data 
#iris.data->150rows*4columns
#iris.feature_names->sets the column headers to the names like 'sepal length(cm)' etc
df['species']= iris.target #adds a new column called 'species' to the DataFrame
#iris.target is a list of integers:0,1,2 representing each species
df['species']=df['species'].map({i:name for i,name in enumerate(iris.target_names)}) #convert the numeric species values to actual names
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
#Histogram for Sepal Length
plt.figure(figsize=(8, 5))
plt.hist(df['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
#plt.hist()->This is a histogram function from Matplotlib's pyplot module(plt).
#bins=20 -> Specifies the no. of bins(intervals) to divide the range of sepal lengths into. eg:if sepal lengths range from 4 to 8 cm, 20 bins would divide that range into 20 small intervals (each~0.2 cm wide)
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Line Plot
plt.plot(df['sepal length (cm)'], label='Sepal Length') #This creates a line plot of sepal length values
plt.plot(df['sepal width (cm)'], label='Sepal Width') #This creates a line plot of sepal width values
plt.legend() #Displays a legend that tells us which line corresponds to which feature.
#it uses the labels we defined in the labels=arguments
plt.title("Line Plot of Sepal Length & Width")
plt.xlabel("Individual Flower Sample")
plt.ylabel("Sepal length and Width")
plt.show()
```

```python
#Visualizing with Seaborn
#Pairplot
sns.pairplot(df, hue='species', palette='husl')
#This generate a grid of scatter plots for every pair of numerical features in our DataFrame df.
#husl stands for 'Hue, Saturation, Lightness' 
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
#Heatmap of Correlation
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
#it returns a table showing how strongly related each pair of features is , using Pearson correlation coefficient
plt.title("Correlation Heatmap")
plt.show()
```

```python
#box plot
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.title("Boxplot of Petal Length by Species")
plt.show()
```

```python
#Violin plot
plt.figure(figsize=(10,6))
sns.violinplot(x='species', y='petal width (cm)', data=df)
plt.title("Violin Plot of Petal Width by Species")
plt.show()
```
