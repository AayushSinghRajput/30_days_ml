```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
#Standardize
X_std = StandardScaler().fit_transform(X)
```

```python
#Apply PCA 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)
```

```python
#Visualize
data_pca = pd.DataFrame(X_pca,columns=['PC1','PC2'])
data_pca['target']=y
sns.scatterplot(data=data_pca,x='PC1',y='PC2',hue='target')
plt.title('PCA - Iris Dataset')
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
#Apply t-SNE
tsne = TSNE(n_components=2,perplexity=30,random_state=42)
X_tsne = tsne.fit_transform(X_std)
```

```python
data_tsne = pd.DataFrame(X_tsne,columns=['Dim1','Dim2'])
data_tsne['target'] = y
sns.scatterplot(data=data_tsne,x="Dim1",y="Dim2",hue="target")
plt.title('t-SNE - Iris Dataset')
plt.show()
```
